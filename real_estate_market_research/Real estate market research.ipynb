Чтобы корректно провести анализ факторов, влияющих на стоимость объектов недвижимости, нужно в первую очередь устранить проблемы в данных. Также для исследования могут понадобиться дополнительные расчетные параметры на основе тех, что имеем в таблице (например, стоимость квадратного метра, соотношение жилой и общей площади, категоризация по этажам).

## Предобработка данных

Переименование столбцов

```python
data = data.rename(columns={'cityCenters_nearest': 'city_centers_nearest', 'parks_around3000': 'parks_around_3000',
                            'ponds_around3000': 'ponds_around_3000'})
data.columns
```

Посчитаем сколько в таблице пропущенных значений

```python
data.isna().sum()
```

Проверяем таблицу на явные дубликаты и `locality_name` на неявные дубликаты

```python
print('Найдено дубликатов:', data.duplicated().sum())
data['locality_name'].unique()
```

Видим, что есть одинаковые названия с разницей в одну букву (например, "посёлок Пансионат Зелёный Бор" и "поселок Пансионат Зелёный Бор"). Есть названия "посёлок городского типа", а есть "городской посёлок" (например, "городской посёлок Мга" и посёлок городского типа Мга"). Исправим это, пусть будет "посёлок" везде, где "поселок". И везде единообразно "посёлок городского типа", вместо "городской посёлок".

```python
data["locality_name"]= data["locality_name"].str.replace("поселок", "посёлок")
data["locality_name"]= data["locality_name"].str.replace("городской посёлок", "посёлок городского типа")
```

### Заполнение пропусков
Почти во всех  столбцах с пропусками находятся количественные переменные, часть из них можно заполнить характерными значениями. Для заполнения пропусков в данных столбцов `ceiling_height`, `floors_total`, `airports_nearest`, `city_centers_nearest`, `parks_nearest`, `ponds_nearest`, `parks_around_3000`, `ponds_around_3000` будем использовать медиану с разбивкой по населенному пункту. Предполагаю, что это не исказит исследование, в этих столбцах не самые важные показатели, и они в среднем могут быть связаны с населенным пунктом. Для заполнения пропусков в данных столбцов `living_area` и `kitchen_area`возьмем медиану с разбивкой по количеству комнат.

Для некоторых пропущенных значений можно предположить логичную замену, например, если человек не указал число балконов (`balcony`), скорее всего их нет. Такие пропуски заменим на 0. В столбце `is_apartment` очень много незаполненных значений, скорее всего это означает, что квартира не является апартаментами, и можно присвоить значение 'False'. 

В столбце `locality_name` всего 49 пропусков. Населенный пункт и достоверность данных по нему очень важны для исследования. Считаю, что эти строки можно удалить, на выборку это не повлиет. В столбце `days_exposition` пропуски решил оставить.

```python
data = data.dropna(subset=['locality_name'])
```

```python
data['balcony'] = data['balcony'].fillna(0)
```

```python
data['is_apartment'] = data['is_apartment'].fillna('False')
```

```python
# Функция для замены пропущенных значений на медиану с разбивкой по населенному пункту
def median_by_locality(column):
    for locality_name in data['locality_name'].unique():
        median = data.loc[data['locality_name'] == locality_name, column].median()
        data.loc[(data[column].isna()) & (data['locality_name'] == locality_name), column] = median
    print(f"Пропуски в столбце {column} после обработки:", data[column].isna().sum())
```

```python
columns_with_gaps = ['ceiling_height', 'floors_total', 'airports_nearest', 'city_centers_nearest', 'parks_nearest',
                    'ponds_nearest', 'parks_around_3000', 'ponds_around_3000']
for column in columns_with_gaps:
    median_by_locality(column)
```

Очевидно, для некоторых населенных пунктов вообще нет данных в нужных столбцах, поэтому и медианы там получаются пустыми. Заполним медианным значением по всему столбцу оставшиеся пропуски в `ceiling_height`. В остальных столбцах пропуски решил оставить, так как больше не вижу подходящего значения для заполнения.

```python
data.loc[data['ceiling_height'].isna(), 'ceiling_height'] = data['ceiling_height'].median()
```

Используем количество комнат в качестве категории для заполнения пропусков по метражу кухонь и жилой площади. Предполагая, что существуют какие-то стандарты строительства.

```python
for room in data['rooms'].unique():
    median_livarea_by_room = data.loc[data['rooms'] == room, 'living_area'].median()
    median_kitarea_by_room = data.loc[data['rooms'] == room, 'kitchen_area'].median()
    data.loc[(data['living_area'].isna()) & (data['rooms'] == room), 'living_area'] = median_livarea_by_room
    data.loc[(data['kitchen_area'].isna()) & (data['rooms'] == room), 'kitchen_area'] = median_kitarea_by_room
print("Пропуски в столбце living_area после обработки:", data['living_area'].isna().sum())
print("Пропуски в столбце kitchen_area после обработки:", data['kitchen_area'].isna().sum())
```

Осталось 197 пустых значений в столбце `kitchen_area`. Убедимся, что это строки, где указано 0 комнат. Возможно, это продавались комнаты в общежитии, где нет своих кухонь.

```python
len(data.loc[(data['kitchen_area'].isna()) & (data['rooms'] == 0)])
```

Заменим эти значения на 0 и посмотрим, что получилось.

```python
data['kitchen_area'] = data['kitchen_area'].fillna(0)
data.isna().sum()
```

### Замена типа данных

Меняем тип данных в тех столбцах, где однозначно бывают только целые значения (число этажей в доме, количество балконов), а также в столбцах `is_apartment`и `first_day_exposition`.

```python
DAY_FORMAT = '%Y.%m.%dT%H:%M:%S'
data['is_apartment'] = data['is_apartment'].astype('bool')
data['floors_total'] = data['floors_total'].astype('int')
data['balcony'] = data['balcony'].astype('int')
data['first_day_exposition'] = pd.to_datetime(
    data['first_day_exposition'], format=DAY_FORMAT)
data.info()
```

Заголовки и неявные дубликаты поправили. Пропущенные значения заменили на подходящие значения там, где это было уместно. Теоретически можно было бы еще избавиться от пропусков в столбце `days_exposition`, но по условию это должно быть количество дней до снятия публикации. Кажется, что лучше оставить значения Nan в этом столбце, что будет означать - объявление еще не снято, на момент получения данных. Тип данных поменяли у 4-ех столбцов.

## Расчёты и добавление результатов в таблицу

Добавляем цену квадртаного метра в таблицу с округлением до двух знаков после запятой:

```python
data['square_meter_price'] = round(data['last_price']/data['total_area'], 2)
```

Добавляем день недели, месяц и год публикации объявления:

```python
data['year_exposition'] = data['first_day_exposition'].dt.year
data['month_exposition'] = data['first_day_exposition'].dt.month
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
```

Добавляем категоризацию по этажу квартиры: варианты — первый, последний, другой.

```python
# функция для распределния объектов по категориям этажности
def floor_group(row):
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'

# добавляем в таблицу столбец с категорией по значению этажа объекта
data['floor_group'] = data.apply(floor_group, axis = 1)

display(data['floor_group'].value_counts())
```

Добавляем в таблицу соотношение жилой и общей площади, а также отношение площади кухни к общей площади (с округлением):

```python
data['ratio_living_total'] = round(data['living_area']/data['total_area'], 2)
data['ratio_kitchen_total'] = round(data['kitchen_area']/data['total_area'], 2)
data.head()
```

### Анализ редких и выбивающихся значений

На гистограммах в предыдущих пунктах исследования не все значения видны, за счет заданного диапазона на многих было не видно выбивающихся значений. Изучим распределения на диаграммах размаха и найдем выбросы в площади объектов, цене, числе комнат, высоте потолков и времени жизни объявлений.

```python
columns_unusual_value = {'total_area':'Общая площадь, кв.м.', 'last_price':'Цена объекта, 100 млн.руб.',
                         'rooms':'Количество комнат, шт.', 'ceiling_height':'Высота потолков, м.',
                         'days_exposition':'Количество дней до продажи'}
# вывод диаграмм размаха через цикл по словарю
for column, rus_name in columns_unusual_value.items():
    data.boxplot(column=[column])
    plt.ylabel(rus_name)
    plt.title(f'Анализ выбросов на диаграмме размаха для столбца [{column}]');
    plt.show()
```

На всех представленных диаграммах размаха выбросами (выбивающимися значениями) можно считать все, что выше верхнего уса или 1.5 межквартильного размаха (IQR) + Q3. Посчитаем эти значения для всех параметров (для расчета IQR возьмем значения 1-го и 3-квартиля из таблицы полученной методом describe в разделе изучения данных):

```python
total_area_blowout = (69.9 - 40) * 1.5 + 69.9
last_price_blowout = (6800000 - 3400000) * 1.5 + 6800000
rooms_blowout = (3 - 1) * 1.5 + 3
ceiling_height_blowout = (2.8 - 2.52) * 1.5 + 2.8
days_exposition_blowout = (232 - 45) * 1.5 + 232
# делаю словарь для выбросов
blowout_dict = {'total_area': total_area_blowout, 'last_price': last_price_blowout,
                'rooms': rooms_blowout, 'ceiling_height': ceiling_height_blowout, 
                'days_exposition': days_exposition_blowout}
# вывожу полученные значения через цикл
for i, j in blowout_dict.items():
    print(f"Выбивающиеся значения для {i} начинаются от:", round(j, 2))
```

Исключаем выбросы из данных с помощью фильтров сделанных с помощью логической индексации и цикла:

```python
for i, j in blowout_dict.items():
    data = data.loc[data[i] <= j] 
```

Методом describe проверяем, что все получилось:

```python
data.describe(include='all').T
```
